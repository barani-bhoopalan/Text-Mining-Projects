#### Basic Text Transformaion

```
#NLTK is a suite of libraries and programs for symbolic and statistical natural language processing for English written in the Python programming language
import nltk

#The NLTK corpus is a massive dump of all kinds of natural language data sets
from nltk.corpus import gutenberg

#to preety print data
from pprint import pprint
```

```
#Get the list of books available in Gutenberg field
nltk.download('gutenberg')
nltk.corpus.gutenberg.fileids()
```

#####A. READ DATA 

```
hamlet = gutenberg.raw(fileids='shakespeare-hamlet.txt')
```

#####B. TOTAL CHARACTERS (Letter by Letter)

```
print(len(hamlet))
```

#####C. TOKENIZE WORDWISE

```
#punkt library for tokenizing
nltk.download('punkt')
```

```
#split long text into words.
tokens = nltk.word_tokenize(hamlet)

## Tokenized type
type(tokens)

##Total words
len(tokens)
```

```
print(tokens)
```

#####D. REMOVE SPECIAL CHAR 

```
#Regex expression
import re
```

```
#import string -  "\w" any word character" 
#re.sub(pattern, repl, string) ---> regex funct
# If it is word--> then make it to lowercase ; else replace it with empty string
normalized_tokens = [re.sub(r'\W+', '', token.lower()) for token in tokens]

##Total words
len(normalized_tokens)
```

```
print(normalized_tokens)
```

#####E. REMOVE SINGLE CHAR (mostly punctuation)

```
#Add only words that have length > 1
normalized_tokens = [normalized_tokens for normalized_tokens in normalized_tokens if len(normalized_tokens)>1]

#Total words after removing single char
len(normalized_tokens)
```

#####F. REMOVE NUMBERS

```
#Add only if the wpord is not a number
normalized_tokens = [normalized_tokens for normalized_tokens in normalized_tokens if not normalized_tokens.isnumeric()]

#Total words after removing numbers
len(normalized_tokens)
```

#####F. REMOVE STOPWORDS

```
from nltk.corpus import stopwords

# stop word is a commonly used word (such as "the") that a search engine has been programmed to ignore
nltk.download('stopwords')
```

```
#Set the Stopword language to ENGLISH
all_stopwords = set(stopwords.words('english'))
```

```
#Add only words that doesnot belong to the stopword list
normalized_tokens = [normalized_tokens for normalized_tokens in normalized_tokens if normalized_tokens not in all_stopwords]

#Total words after removing stopwords
len(normalized_tokens)
```

#####G. FREQUENCY DISTRIBUTION

```
#NLTK package to find frequency distribution of words
frq_dist = nltk.FreqDist(normalized_tokens)

#Top 10 words and its frequency
frq_dist.most_common(10)
```

#####H. STEMMING

```
# stemming --> stemming is the process of reducing inflected words to their word stem, base or root form
from nltk.stem import PorterStemmer
```

```
ps = PorterStemmer()

for normalized_tokens in normalized_tokens:
  print(normalized_tokens + ":" + ps.stem(normalized_tokens))
```
